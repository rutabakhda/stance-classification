```python
import os
import pandas as pd
import re
```

```python
folder_path = "../data/sample-sbm"
all_files = os.listdir(folder_path)
```

```python
list_topic = ["Saudi Arabia", "Saudi Arabia", "Racism", "Prostitution", "Prostitution", "Poverty", "Poverty", "Pornography", "Child Pornography", "Marijuana", "Marijuana", "Gun Control"]
```

```python
#df = pd.DataFrame(columns=['id','sentence','unitType','topic'])
```

```python
print(all_files)
```

```python
all_files.sort()
print(all_files)
i=0
all_labels = []
for filename in all_files:
#     filename = all_files[i]
    print(filename)
    topic = list_topic[i]
    i=i+1
    file_path = os.path.join(folder_path, filename)
    with open(file_path) as f_in:
        lines = (line.rstrip() for line in f_in) # All lines including the blank ones
        lines = (line for line in lines if line) # Non-blank lines
        
        for line in lines:
            print(line)
            pattern_premise = "[nonconclusion]"
            pattern_conclusion = "[conclusion]"
            is_premise = pattern_premise in line
            is_conclusion = pattern_conclusion in line
            sentence=""
            unit_type=""
            if is_premise:
                unit_type = "premise"
                sentence = line.replace(pattern_premise, "")
            if is_conclusion:
                unit_type = "conclusion"
                sentence = line.replace(pattern_conclusion, "")
            if len(unit_type)>0 and len(sentence)>0:
                print(sentence)
                label = {'id':str(i),'content':sentence,"unitType":unit_type,"topic":topic}
                all_labels.append(label)
            
df = pd.DataFrame(all_labels)
```
