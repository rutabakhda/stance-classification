```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import os
```

# Full Data

```python
pathToData = "/home/weci2587/projects/wstud-visit-the-dome-ss19/data/demo/processed/webis-debate-16/full"
df = pd.DataFrame()
df["Document"] = np.nan
df["UnitsTotal"] = np.nan
df["Argumentative"] = np.nan
df["Non-Argumentative"] = np.nan
dirs = dir_list = next(os.walk(pathToData))[1]

for dir in dirs:
    documentPath = pathToData + "/" + dir
    total = 0
    argumentative = 0
    nonArgumentative = 0
    files = os.listdir(documentPath)
    for file in files:
        total += 1
        f = open(documentPath + "/" + file, "r")
        content = f.read()
        if "unitType=\"Non-Argumentative\"" in content:
            nonArgumentative += 1
        elif "unitType=\"Argumentative\"" in content:
            argumentative +=1 
        f.close()
    df = df.append({"Document": dir, "UnitsTotal": total, "Argumentative": argumentative, "Non-Argumentative": nonArgumentative},\
                  ignore_index = True)
df
```

```python
df.describe().T
```

```python
df.sum()
```
